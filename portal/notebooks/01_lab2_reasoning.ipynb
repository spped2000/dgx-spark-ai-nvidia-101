# LAB 2 — โมเดล 'คิดให้ดู' vs 'ตอบเลย' (Reasoning)
โมเดลตัวเดียวกัน สั่งให้ **คิดทีละขั้น** หรือ **ตอบเลย** ได้ — กด `Shift+Enter` ทีละเซลล์

```python
import os, time
from openai import OpenAI
client = OpenAI(base_url=os.environ['OPENAI_BASE_URL'], api_key='EMPTY'); MODEL=os.environ['MODEL']
Q = 'รถออก 09:00 ความเร็ว 60 กม./ชม. ระยะ 150 กม. ถึงกี่โมง? แสดงวิธีคิด'
```

```python
# 🟢 ตอบเลย (thinking OFF)
t=time.time()
r=client.chat.completions.create(model=MODEL,max_tokens=400,messages=[{'role':'system','content':'detailed thinking off'},{'role':'user','content':Q}])
print(f'tokens={r.usage.completion_tokens}  เวลา={time.time()-t:.1f}s\n')
print(r.choices[0].message.content)
```

```python
# 🔵 คิดทีละขั้น (thinking ON) — ช้ากว่า/ยาวกว่า
t=time.time()
r=client.chat.completions.create(model=MODEL,max_tokens=600,messages=[{'role':'system','content':'detailed thinking on'},{'role':'user','content':Q}])
print(f'tokens={r.usage.completion_tokens}  เวลา={time.time()-t:.1f}s\n')
print(r.choices[0].message.content)
```

**สังเกต:** ON สร้าง token เยอะ/ช้ากว่า (โชว์วิธีคิด) · OFF สั้น/เร็ว → **ฉลาดขึ้น = แพงขึ้น** เลือกใช้ให้เหมาะงาน
