# 👋 ยินดีต้อนรับสู่ DGX Spark Workshop
นี่คือ **sandbox ส่วนตัวของคุณ** (container แยกของคุณเอง) บน DGX Spark เครื่องเดียวที่ทั้งห้องแชร์กัน

**ไม่ต้องเขียนโค้ด** — แค่กดเล่นทีละเซลล์ (กด `Shift+Enter`)

```python
import os
from openai import OpenAI
client = OpenAI(base_url=os.environ.get('OPENAI_BASE_URL'), api_key='EMPTY')
MODEL = os.environ.get('MODEL', 'nemotron-nano')
print('เชื่อมต่อ DGX Spark ผ่านโมเดลกลาง:', MODEL, '@', os.environ.get('OPENAI_BASE_URL'))
```

```python
# 🤖 ลองคุยกับ AI ที่รันบน DGX Spark — แก้ข้อความแล้วกด Shift+Enter ได้เลย
prompt = 'แนะนำ DGX Spark แบบสั้นๆ ให้คนเริ่มต้นเข้าใจ'
r = client.chat.completions.create(model=MODEL, messages=[{'role':'user','content':prompt}], max_tokens=150)
print(r.choices[0].message.content)
```

✅ ถ้าเห็นคำตอบด้านบน = sandbox ของคุณเชื่อมกับโมเดลกลางบน DGX Spark สำเร็จแล้ว!
