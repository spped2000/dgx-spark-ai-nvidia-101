# LAB 5 — Secure & Sandbox (ดูผู้สอนสาธิตสด 👀)
LAB นี้ **ผู้สอนสาธิตสด** (OpenShell sandbox เป็น ALPHA — ไม่ได้แยกให้ทุกคนคนละตัว) — โน้ตบุ๊กนี้คือสรุปสิ่งที่จะได้เห็น

## จะเห็นอะไร
1. agent ในกล่อง (sandbox) ลอง `curl https://api.github.com/zen` → ✅ **ALLOWED**
2. ลอง `POST` สร้าง issue → ❌ **BLOCKED** (`policy_denied`, ชั้น L7) — *เขียนไม่ได้!*
3. ลองออกเน็ตนอก policy (`example.com`) → ❌ **BLOCKED** — *กันข้อมูลรั่ว*
4. ผู้สอน **เพิ่มกฎสด** (`policy update --add-allow`) → POST ผ่านทันที (ไม่ต้อง restart)

## บทเรียน
**deny-by-default → เปิดสิทธิ์ทีละกฎ → บันทึกทุก action (audit)** = ปล่อย AI รันคำสั่งได้อย่างปลอดภัยและควบคุมได้

```python
# (ตัวเลือก) ลองคุยกับ AI ว่า sandbox/policy คืออะไร ผ่านโมเดลกลาง
import os
from openai import OpenAI
client = OpenAI(base_url=os.environ['OPENAI_BASE_URL'], api_key='EMPTY')
r=client.chat.completions.create(model=os.environ['MODEL'],max_tokens=120,messages=[{'role':'user','content':'อธิบายสั้นๆ ว่า agent sandbox แบบ deny-by-default ช่วยความปลอดภัยยังไง'}])
print(r.choices[0].message.content)
```

> นโยบาย (policy.yaml) ที่ใช้: อนุญาต GitHub API แบบ **read-only** ผ่าน `curl` เท่านั้น — อย่างอื่นบล็อกหมด
> ผู้สอนรันเดโมจาก `tools/phoenix_sandbox_live.py` + `openshell term` (ดูบนจอใหญ่)
